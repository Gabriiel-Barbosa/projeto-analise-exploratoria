# Análise de Uma Base de Dados de Vendas

## Importando Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn")
```

## Criando o DataFrame

```python
df = pd.read_excel('datasets/AdventureWorks.xlsx')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

## Verificando Valores Nulos 

```python
df.isnull().sum()
```

```python
df.sample(25)
```

```python
df['Produto'].unique()
```

## Calculando a Receita Total 

```python
receita_total = df['Valor Venda'].sum()
```

```python
receita_total
```

## Custo Total 

```python
df['Custo Total'] = df ['Custo Unitário'].mul(df['Quantidade'])
```

```python
custo_total = round (df['Custo Total'],2).sum()
```

```python
custo_total
```

```python
df.sample(10)
```

## Lucro 

```python
df['Lucro'] = df['Valor Venda'] - df['Custo Total']
```

```python
lucro_total = round(df['Lucro'], 2).sum()
```

```python
lucro_total
```

```python
df.sample(10)
```

## Total de Dias Para Enviar o Produto 

```python
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df
```

## Média de Envio Por Marca

```python
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
media_envio_marca = df.groupby("Marca")["Tempo_envio"].mean()
```

```python
round(media_envio_marca, 2)
```

## Lucro Por Ano e Por Marca

### Agrupando Por Ano e Por Marca

```python
 pd.options.display.float_format = '{:20,.2f}'.format
```

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
```

```python
lucro_ano
```

## ## Total de Produtos Vendidos

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

## Análises Gráficas 

### Lucro x Ano

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano", color = '#483D8B')
plt.xlabel("Ano")
plt.ylabel("Receita");
```

### Produtos Vendidos

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos", color = "#483D8B")
plt.xlabel("Total")
plt.ylabel("Produto");
plt.xticks(rotation='horizontal');
```

## Análise de Vendas do  Ano - 2019

```python
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

### Lucro x Mês (2019)

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês", color = "#483D8B")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

### Lucro x Marca

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca", color = "#483D8B")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

## Estatísticas 

### Estátisicas Gerais

```python
df_estatisticas = round(df["Tempo_envio"].describe(),2)
```

```python
plotdf_estatisticas =df_estatisticas.plot.bar(title="Estatísticas", color = "#483D8B")
for i in plotdf_estatisticas.patches:
    plotdf_estatisticas.annotate(i.get_height(),
                  (i.get_x() + i.get_width() / 2, i.get_height()),
                  ha='center',
                  va='baseline',
                  fontsize=12,
                  color='black',
                  xytext=(0, 1),
                  textcoords='offset points')
plt.xticks(rotation='horizontal');
```
