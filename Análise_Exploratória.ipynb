## Calculando a Receita Total 

```python
receita_total = df['Valor Venda'].sum()
```

```python
receita_total
```

## Custo Total 

```python
df['Custo Total'] = df ['Custo Unitário'].mul(df['Quantidade'])
```

```python
custo_total = round (df['Custo Total'],2).sum()
```

```python
custo_total
```

```python
df.sample(10)
```

## Lucro 

```python
df['Lucro'] = df['Valor Venda'] - df['Custo Total']
```

```python
lucro_total = round(df['Lucro'], 2).sum()
```

```python
lucro_total
```
